# Análise Exploratória de Dados (EDA)
### Conjunto de dados: vehicles_us.csv

Este notebook realiza uma análise exploratória básica dos anúncios de veículos usados
nos Estados Unidos. Utilizaremos Plotly Express para criar visualizações interativas
que nos ajudem a compreender melhor o conjunto de dados.

```python
import pandas as pd
import plotly.express as px
```

```python
# Ler o conjunto de dados
car_data = pd.read_csv('projeto/vehicles.csv')
car_data.head()
```

```python
# Histograma da quilometragem
fig1 = px.histogram(car_data, x='odometer', nbins=50,
                    title='Distribuição da quilometragem dos veículos')
fig1.show()
```

```python
# Histograma dos preços
fig2 = px.histogram(car_data, x='price', nbins=50,
                    title='Distribuição dos preços dos veículos')
fig2.show()
```

```python
# Dispersão entre preço e quilometragem
fig3 = px.scatter(car_data, x='odometer', y='price',
                  title='Relação entre preço e quilometragem')
fig3.show()
```

```python
# Boxplot do preço por tipo de combustível
fig4 = px.box(car_data, x='fuel', y='price',
              title='Preço dos veículos por tipo de combustível')
fig4.show()
```

```python
# Número de anúncios por tipo de veículo
fig5 = px.bar(car_data['type'].value_counts().reset_index(),
              x='index', y='type',
              labels={'index': 'Tipo de veículo', 'type': 'Quantidade'},
              title='Número de anúncios por tipo de veículo')
fig5.show()
```

### ✅ Conclusões iniciais:
- A maioria dos veículos tem quilometragem entre 50.000 e 150.000 milhas.
- A distribuição de preços é assimétrica, com muitos carros mais baratos.
- Há correlação negativa entre preço e quilometragem.
- O tipo de combustível e o tipo de veículo parecem influenciar o preço.

Essas observações ajudam a guiar as interações no aplicativo Streamlit.
